```python
# importing libraries
from langchain_community.retrievers import WikipediaRetriever
```

```python
# Creating Retriever
retriever = WikipediaRetriever(
    top_k_results=2,
    lang='en'
)
```

```python
docs = retriever.invoke("what are the applications of RAG")
```
