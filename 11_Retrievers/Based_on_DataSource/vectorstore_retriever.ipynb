```python
# Importing necessary libraires
from langchain_huggingface import HuggingFaceEmbeddings
from langchain.vectorstores import Chroma
from langchain_core.documents import Document
```

```python
# Creating embedding model
import os
os.environ['HF_HOME'] = 'D:/huggingface_cache'

embeddings = HuggingFaceEmbeddings(model = "sentence-transformers/all-MiniLM-L6-v2")
```

```python
# Creating Documents

doc1 = Document(
        page_content="LangChain provides a unified interface to work with large language models (LLMs) from providers like OpenAI, Anthropic, and Hugging Face. It allows developers to easily switch between models and configure parameters such as temperature, max tokens, and streaming responses."
    )

doc2 = Document(
        page_content="Prompt templates in LangChain help structure and reuse prompts efficiently. Developers can define dynamic templates with variables, making it easier to generate consistent and customizable inputs for language models."
    )

doc3 = Document(
        page_content="Chains in LangChain allow multiple components such as prompts, language models, and output parsers to be combined into a sequence of operations. This enables building complex workflows where the output of one step becomes the input of the next."
    )

doc4 = Document(
        page_content="LangChain supports integration with vector stores like Chroma and FAISS to enable retrieval-based applications. Documents are converted into embeddings and stored, allowing similarity search to fetch relevant context for answering queries."
    )

doc5 = Document(
        page_content="Agents in LangChain enable language models to make decisions and interact with external tools such as APIs, calculators, and databases. They can dynamically choose which actions to take based on user input and intermediate reasoning steps."
    )

docs = [doc1, doc2, doc3, doc4, doc5]
```

```python
# Crete Vector Store
vector_store = Chroma.from_documents(
    documents=docs,
    embedding= embeddings,
    collection_name='sample'
)
```

```python
vector_store.get(include=['embeddings', 'metadatas', 'documents'])
```

```python
retriever = vector_store.as_retriever(search_kwargs={'k' : 2})
```

```python
result = retriever.invoke("What is chains in langchain?")
```
