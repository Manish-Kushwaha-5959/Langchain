# Basic on Runnables

```python
from abc import ABC, abstractmethod
```

```python
class Runnabble(ABC):
    
    @abstractmethod
    def invoke(self, input_data):
        pass
```

```python
class NakliLLM(Runnabble):
    def __init__(self):
        pass

    def invoke(self, input_data):
        return {'response' : f'llm response for query : {input_data}'}

    def predict(self, query):
        return {'response' : f'llm response for query : {query}'}
```

```python
class NakliPromptTemplate(Runnabble):
    def __init__(self, template, input_variables):
        self.template = template
        self.input_variables = input_variables

    def invoke(self, input_dict):
        return self.template.format(**input_dict)

    def format(self, input_dict):
        return self.template.format(**input_dict)
```

```python
class Chain(Runnabble):
    def __init__(self, runnable_list):
        self.runnable_list = runnable_list

    def invoke(self, input_data):
        for runnable in self.runnable_list:
            input_data = runnable.invoke(input_data)
        return input_data
```

```python
class NakliStrOuputParser(Runnabble):
    def __init__(self):
        pass
    
    def invoke(self, input_data):
        return input_data['response']
```

```python
prompt1 = NakliPromptTemplate(
    template="tell me a joke on {topic}",
    input_variables=["topic"]
)

prompt2 = NakliPromptTemplate(
    template="Explain me the below joke {response}",
    input_variables=['response']
)
```

```python
model = NakliLLM()
```

```python
parser = NakliStrOuputParser()
```

```python
chain1 = Chain([prompt1, model])
chain2 = Chain([prompt2, model, parser])

final_chain = Chain([chain1, chain2])

final_chain.invoke({'topic' : 'college'})
```
