```python
from langchain_community.tools import tool
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.messages import HumanMessage, AIMessage, SystemMessage
from dotenv import load_dotenv
```

```python
load_dotenv()
model = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
```

```python
@tool
def multiply(a: int, b: int) -> int:
    """"return multiplication of two numbers"""
    return a*b
```

```python
multiply.invoke({'a' : 4, 'b' : 5})
```

```python
llm_with_tool = model.bind_tools([multiply])
```

```python
message = [SystemMessage(content="you are a helpful chat assistant use the provided tool to answer query after getting the result from tool format the answer properly and reponse to the user query")]
query = "multiply 3 by 4 using the provided tool"
message.append(HumanMessage(content=query))
message
```

```python
result = llm_with_tool.invoke(message)
result.tool_calls[0]
```

```python
multiply.invoke(result.tool_calls[0]['args'])
```

```python
message.append(multiply.invoke(result.tool_calls[0]))
message
```

```python
result = llm_with_tool.invoke(message)
result
```
