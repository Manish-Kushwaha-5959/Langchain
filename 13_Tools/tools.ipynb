# Custom Tools

## Method-2 tool decorator

```python
from langchain_community.tools import tool

@tool
def multiply(a: int, b: int) -> int:
    """Multiplies two integers"""
    return a * b

result = multiply.invoke({"a": 2, "b": 3})
print(result)
print()
print(multiply.name)
print(multiply.description)
print(multiply.args)
```

```python
multiply.args_schema.model_json_schema()
```

## Method 2 - Structure tool

```python
from langchain_community.tools import StructuredTool
from pydantic import BaseModel, Field
```

```python
class MultiplyInput(BaseModel):
    a : int = Field(required = True, description = "First number to multiply")
    b : int = Field(required = True, description = "Second number to multiply")
```

```python
def multiply_func(a: int, b: int) -> int:
    """Multiplies two integers"""
    return a * b
```

```python
multiply_tool = StructuredTool.from_function(
    func = multiply_func,
    name = "multiply",
    description = "Multiplies two integers",
    args_schema = MultiplyInput
)
```

```python
multiply_tool.invoke({'a' : 5, 'b' : 10})
```

```python
print(multiply_tool.name)
print(multiply_tool.description)
print(multiply_tool.args)
```

```python
multiply_tool.args_schema.model_json_schema()
```

## Method-3: BaseTool

```python
from langchain_community.tools import BaseTool
from typing import Type
```

```python
class MultiplyInput(BaseModel):
    a : int = Field(required = True, description = "First number to multiply")
    b : int = Field(required = True, description = "Second number to multiply")
```

```python
class MultiplyTool(BaseTool):
  name: str = "Multiply"
  description: str = "Multiplies two integers"
  args_schema: Type[BaseModel] = MultiplyInput

  def _run(self, a: int, b: int) -> int:
    return a * b
```

```python
multiply_tool = MultiplyTool()
multiply_tool.invoke({'a': 2, 'b' : 3})
```

```python
print(multiply_tool.name)
print(multiply_tool.description)
print(multiply_tool.args)
```

```python
multiply_tool.args_schema.model_json_schema()
```
