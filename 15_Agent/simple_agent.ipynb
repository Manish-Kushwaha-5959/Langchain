```python
from langchain_community.tools import tool, DuckDuckGoSearchRun
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.agents import create_react_agent, AgentExecutor
from langchain import hub
import time
import requests
import os
from dotenv import load_dotenv
```

```python
load_dotenv()
model = ChatGoogleGenerativeAI(model="gemini-2.5-flash")

search_tool = DuckDuckGoSearchRun()
```

```python
prompt = hub.pull("hwchase17/react") #pull the standard ReAct agent prompt
prompt
```

```python
agent = create_react_agent(
    llm = model,
    tools = [search_tool],
    prompt = prompt
)
```

```python
agent_executor = AgentExecutor(
    agent = agent,
    tools = [search_tool],
    verbose = True
)
```

```python
response = agent_executor.invoke({"input" : "what is the population of the capital of inida"})
print(response)
```
