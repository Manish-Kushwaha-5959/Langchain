```python
# Importing necessary libraires
from langchain_huggingface import HuggingFaceEmbeddings
from langchain.vectorstores import Chroma
from langchain_core.documents import Document
```

```python
# Creating embedding model
import os
os.environ['HF_HOME'] = 'D:/huggingface_cache'

embeddings = HuggingFaceEmbeddings(model = "sentence-transformers/all-MiniLM-L6-v2")
```

```python
# Create Langchain Documents 

doc1 = Document(
        page_content="Artificial Intelligence (AI) is transforming industries through automation, predictive analytics, and natural language processing. Modern AI systems leverage deep learning models trained on large datasets.",
        metadata={
            "category": "AI"
        }
    )

doc2 = Document(
        page_content="Cloud computing enables scalable infrastructure and services over the internet. Providers like AWS, Azure, and GCP offer IaaS, PaaS, and SaaS solutions for businesses.",
        metadata={
            "category": "Cloud"
        }
    )

doc3 = Document(
        page_content="Blockchain technology provides a decentralized ledger system that ensures transparency and immutability. It is widely used in cryptocurrencies and supply chain management.",
        metadata={
            "category": "Blockchain"
        }
    )

doc4 = Document(
        page_content="Cybersecurity involves protecting systems, networks, and data from digital attacks. Techniques include encryption, firewalls, and intrusion detection systems.",
        metadata={
            "category": "Cybersecurity"
        }
    )

doc5 = Document(
        page_content="The Internet of Things (IoT) connects physical devices to the internet, enabling data exchange and smart automation across homes, industries, and cities.",
        metadata={
            "category": "IoT"
        }
    )

docs = [doc1, doc2, doc3, doc4, doc5]
```

```python
vector_store = Chroma(
    embedding_function = embeddings,
    persist_directory = r'chroma_db',
    collection_name = 'sample'
)
```

```python
# Add documents

vector_store.add_documents(docs)
```

```python
vector_store.get(include=['embeddings', 'documents', 'metadatas'])
```

```python
# Search Documents
vector_store.similarity_search(
    query="companies in cloud technologies",
    k=1
)
```

```python
# Search with similarity score
vector_store.similarity_search_with_score(
    query='companies in cloud technologies',
    k=2
)
```

```python
# meta-data filtering
vector_store.similarity_search_with_score(
    query='',
    filter={'category' : 'AI'}
)
```

```python
# Update document
updated_doc1 = Document(
    page_content="Artificial Intelligence (AI) is widely used in real-world applications such as virtual assistants, recommendation systems, fraud detection, and autonomous vehicles, improving efficiency and decision-making across industries.",
    metadata={"category": "AI"}
)

vector_store.update_document(document_id='b23e023b-0d64-4560-a1ea-76405f8f6bbf', document=updated_doc1)
```
